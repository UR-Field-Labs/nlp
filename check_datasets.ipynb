```python
import pandas as pd
import os
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Define the path to the directory containing your CSV files
csv_directory = '/content/drive/MyDrive/newspapers/'

# List of CSV file names (ensure these names are correct)
csv_files = [
    'Setopati_English_preprocessed.csv',
    'himalayan_times_preprocessed_data.csv',
    'kathmundu_post_preprocessed.csv',
    'Copy of online_khabar_english_preprocessed_data.csv'
]

# List to hold individual DataFrames
dataframes = []

# Read each CSV file into a DataFrame and add it to the list
for file_name in csv_files:
    file_path = os.path.join(csv_directory, file_name)
    if os.path.exists(file_path):
        print(f"Reading {file_path}")
        df = pd.read_csv(file_path)
        if not df.empty:
            dataframes.append(df)
        else:
            print(f"Warning: {file_path} is empty and will be skipped.")
    else:
        print(f"Warning: {file_path} does not exist and will be skipped.")

# Check if we have any DataFrames to concatenate
if dataframes:
    # Concatenate all DataFrames in the list into a single DataFrame
    combined_df = pd.concat(dataframes, ignore_index=True)

    # Save the combined DataFrame to a new CSV file
    output_file_path = os.path.join(csv_directory, 'combined_file.csv')
    combined_df.to_csv(output_file_path, index=False)

    print(f"Concatenation complete. The combined file is saved as '{output_file_path}'.")
else:
    print("No valid CSV files were read. No concatenation performed.")

```
